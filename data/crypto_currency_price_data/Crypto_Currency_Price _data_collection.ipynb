```python
import requests
import pandas as pd
```

# Collecting last 7 days minute data

```python
# Mon, 05 Mar 2018 02:47:00 GMT
toTime = 1520218020
timeGapMins=651*60
url = 'https://min-api.cryptocompare.com/data/histominute?fsym=BTC&tsym=USD&limit=650&toTs= {}'
master_df = pd.DataFrame()
while True:
    # r is a dictionary that contains the json response
    r = requests.get(url.format(toTime)).json()
    data = r['Data']
    if len(data) == 0 :
        print('No data was returned from the API call!')
        r.json()
        break
        
    close =[]
    high =[]
    low = []
    open = []
    time = []
    volumefrom = []
    volumeto = []
    for i in range(len(data)):
        close.append(data[i]['close'])
        high.append(data[i]['high'])
        low.append(data[i]['low'])
        open.append(data[i]['open'])
        time.append(data[i]['time'])
        volumefrom.append(data[i]['volumefrom'])
        volumeto.append(data[i]['volumeto'])

    master_df = pd.concat([master_df,pd.DataFrame(
        {
     'close': close,
     'high': high,
     'low': low,
     'open': open,
     'time': time,
     'volumefrom': volumefrom,
     'volumeto': volumeto
        }
    )    
    ])
    toTime = toTime-timeGapMins
    if len(data)<651:
        break
```

Note that there are some index numbers duplicated due to the pd.concat() method in the for loop (see above). This may cause an error if we directly save the master_df as a local file. Let's reset the index first.

```python
master_df = master_df.sort_values(by='time', ascending=False)
master_df.index = range(1,len(master_df)+1)
master_df.head(5)
```

# Saving the minute data in csv file

```python
master_df.to_csv('minute_data_till_05_Mar_2018_02_47_00.csv')
```

# Collecting last two weeks hourly data

```python
#Monday, March 5, 2018 3:00:00 AM
toTime_hourly = 1520218800
hourly_url = 'https://min-api.cryptocompare.com/data/histohour?fsym=BTC&tsym=USD&limit=335&toTs={}'
r = requests.get(hourly_url.format(toTime_hourly)).json()
data = r['Data']
```

```python
master_df = pd.DataFrame()
close =[]
high =[]
low = []
open = []
time = []
volumefrom = []
volumeto = []
for i in range(len(data)):
    close.append(data[i]['close'])
    high.append(data[i]['high'])
    low.append(data[i]['low'])
    open.append(data[i]['open'])
    time.append(data[i]['time'])
    volumefrom.append(data[i]['volumefrom'])
    volumeto.append(data[i]['volumeto'])

master_df = pd.concat([master_df,pd.DataFrame({'close': close,
 'high': high,
 'low': low,
 'open': open,
 'time': time,
 'volumefrom': volumefrom,
 'volumeto': volumeto})    
])
master_df.head(5)
```

```python
master_df =  master_df.sort_values(by='time', ascending=False)
master_df.index = range(1,len(master_df)+1)
master_df.head(5)
```

```python
master_df.info()
```

# Writing hourly data to csv file

```python
master_df.to_csv('hourly_data_till_05_Mar_2018_03_00_00.csv')
```
